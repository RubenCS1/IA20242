```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import accuracy_score, classification_report

#Preparar Datos
#Dividir características y etiqueta objetivo
X = Data.drop(columns=['RENDIMIENTO_GLOBAL'])
y = Data['RENDIMIENTO_GLOBAL']

#Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Convertir la columna objetivo a categorías para la red neuronal
#Dado que generaba error, se ajustan a 0,1,2 y 3
y_categorical = to_categorical(y - 1)

#Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_categorical, test_size=0.3, random_state=42)

#Calcular pesos de clase para manejar desequilibrio
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(Data['RENDIMIENTO_GLOBAL']),
    y=Data['RENDIMIENTO_GLOBAL']
)
class_weights_dict = dict(enumerate(class_weights))

# Configurar la Red Neuronal Mejorada
model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(512, kernel_regularizer='l2'),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dropout(0.4),
    Dense(256, kernel_regularizer='l2'),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dropout(0.3),
    Dense(128, kernel_regularizer='l2'),
    BatchNormalization(),
    LeakyReLU(alpha=0.1),
    Dense(4, activation='softmax')
])

#Compilar el modelo con Adam y un learning rate dinámico
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

#Configurar callbacks: EarlyStopping y ReduceLROnPlateau
early_stopping = EarlyStopping(
    monitor='val_loss',
# Detener si no mejora después de 20 épocas
    patience=20,
    restore_best_weights=True
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-5
)

#Entrenar el modelo
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=512,
    class_weight=class_weights_dict,  # Usar pesos de clase
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)


#Evaluar en el conjunto de prueba
y_pred_nn = model.predict(X_test)
# Convertir a etiquetas originales [1, 2, 3, 4]
y_pred_labels = y_pred_nn.argmax(axis=1) + 1
y_test_labels = y_test.argmax(axis=1) + 1

#Calcular métricas
accuracy_nn = accuracy_score(y_test_labels, y_pred_labels)
report_nn = classification_report(y_test_labels, y_pred_labels)

print("Precisión del modelo mejorado (Red Neuronal):", accuracy_nn)
print("Reporte de clasificación:\n", report_nn)

# Guardar el Modelo
model.save("modelo_red_neuronal_mejorado_avanzado.h5")
print("Modelo guardado como 'modelo_red_neuronal_mejorado_avanzado.h5'.")
```
